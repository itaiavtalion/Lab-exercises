```python
import requests

#get info about the latest launches
# Define the API URL
latest_launch_url = "https://api.spacexdata.com/v4/launches/latest"

# Make a GET request 
response = requests.get(latest_launch_url)

latest_launch = response.json()
print("Latest launch name:", latest_launch["name"])
print("Latest launch date:", latest_launch["date_utc"])
print("Latest launch rocket name:", latest_launch["rocket"])
```

```python
#get info about all rockets 
rockets_url = "https://api.spacexdata.com/v4/rockets"


response = requests.get(rockets_url)


rockets = response.json()
for rocket in rockets:
    print("Rocket name:", rocket["name"])
    print("Rocket height (meters):", rocket["height"]["meters"])
    print("Rocket mass (kg):", rocket["mass"]["kg"])
    print("")
```

```python
#Get information about all past launches
past_launches_url = "https://api.spacexdata.com/v4/launches/past"

response = requests.get(past_launches_url)


past_launches = response.json()
for launch in past_launches:
    print("Launch name:", launch["name"])
    print("Launch date:", launch["date_utc"])
    print("Rocket name:", launch["rocket"])
    print("")
```
